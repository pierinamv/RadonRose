## Tratando los Fondos G2 y TC

```python
B_G2_Fondo_Mean = B_LR_L.iloc[42::,:]
B_G2_Fondo_Mean = B_G2_Fondo_Mean.copy()
B_G2_Fondo_Mean.rename(columns={'B_tr/cm2':'tr/cm2','B_uncert':'uncert'},inplace=True)

C_G2_Fondo_Mean = C_LR_L.iloc[42::,:]
C_G2_Fondo_Mean = C_G2_Fondo_Mean.copy()
C_G2_Fondo_Mean.rename(columns={'C_tr/cm2':'tr/cm2','C_uncert':'uncert'},inplace=True)

D_G2_Fondo_Mean = D_LR_L.iloc[42::,:]
D_G2_Fondo_Mean = D_G2_Fondo_Mean.copy()
D_G2_Fondo_Mean.rename(columns={'D_tr/cm2':'tr/cm2','D_uncert':'uncert'},inplace=True)

E_G2_Fondo_Mean = E_LR_L.iloc[42::,:]
E_G2_Fondo_Mean = E_G2_Fondo_Mean.copy()
E_G2_Fondo_Mean.rename(columns={'E_tr/cm2':'tr/cm2','E_uncert':'uncert'},inplace=True)
```

```python
B_G2_Fondo_Mean 
```

```python
G2_Fondos=pd.concat([B_G2_Fondo_Mean,C_G2_Fondo_Mean,D_G2_Fondo_Mean,E_G2_Fondo_Mean])
G2_Fondos
```

```python
def incertidumbre_media(col):
    n = len(col)
    suma_cuadrados = np.sum(np.square(col))
    incertidumbre_promedio = np.sqrt(suma_cuadrados) / n

    return incertidumbre_promedio
```

```python
G2_Fondo_mean = G2_Fondos.iloc[:,[1]].mean().values[0]
G2_Fondo_mean
```

```python
G2_Fondo_uncert = G2_Fondos.iloc[:,[2]].apply(incertidumbre_media).values[0]
G2_Fondo_uncert
```

```python
#B
B_TC1_Fondo_Mean = B_LR_L.iloc[24:30,:].iloc[::2,:]
B_TC1_Fondo_Mean = B_TC1_Fondo_Mean.copy()
B_TC1_Fondo_Mean.rename(columns={'B_tr/cm2':'tr/cm2','B_uncert':'uncert'},inplace=True)

B_TC2_Fondo_Mean = B_LR_L.iloc[24:30,:].iloc[1::2,:]
B_TC2_Fondo_Mean = B_TC2_Fondo_Mean.copy()
B_TC2_Fondo_Mean.rename(columns={'B_tr/cm2':'tr/cm2','B_uncert':'uncert'},inplace=True)

#C
C_TC1_Fondo_Mean = C_LR_L.iloc[24:30,:].iloc[::2,:]
C_TC1_Fondo_Mean = C_TC1_Fondo_Mean.copy()
C_TC1_Fondo_Mean.rename(columns={'C_tr/cm2':'tr/cm2','C_uncert':'uncert'},inplace=True)

C_TC2_Fondo_Mean = C_LR_L.iloc[24:30,:].iloc[1::2,:]
C_TC2_Fondo_Mean = C_TC2_Fondo_Mean.copy()
C_TC2_Fondo_Mean.rename(columns={'C_tr/cm2':'tr/cm2','C_uncert':'uncert'},inplace=True)

#D
D_TC1_Fondo_Mean = D_LR_L.iloc[24:30,:].iloc[::2,:]
D_TC1_Fondo_Mean = D_TC1_Fondo_Mean.copy()
D_TC1_Fondo_Mean.rename(columns={'D_tr/cm2':'tr/cm2','D_uncert':'uncert'},inplace=True)

D_TC2_Fondo_Mean = D_LR_L.iloc[24:30,:].iloc[1::2,:]
D_TC2_Fondo_Mean = D_TC2_Fondo_Mean.copy()
D_TC2_Fondo_Mean.rename(columns={'D_tr/cm2':'tr/cm2','D_uncert':'uncert'},inplace=True)

#E
E_TC1_Fondo_Mean = E_LR_L.iloc[24:30,:].iloc[::2,:]
E_TC1_Fondo_Mean = E_TC1_Fondo_Mean.copy()
E_TC1_Fondo_Mean.rename(columns={'E_tr/cm2':'tr/cm2','E_uncert':'uncert'},inplace=True)

E_TC2_Fondo_Mean = E_LR_L.iloc[24:30,:].iloc[1::2,:]
E_TC2_Fondo_Mean = E_TC2_Fondo_Mean.copy()
E_TC2_Fondo_Mean.rename(columns={'E_tr/cm2':'tr/cm2','E_uncert':'uncert'},inplace=True)
```

```python
def incertidumbre_resta(u_x):
    u_y = TC2_Fondo_uncert
    return np.sqrt(u_x**2 + u_y**2)
```

```python
col = ['B_tr/cm2', 'C_tr/cm2', 'D_tr/cm2', 'E_tr/cm2']
TC2_L[col] = TC2_L[col] - TC2_Fondo_mean
G2_L[col] = G2_L[col] - G2_Fondo_mean
```

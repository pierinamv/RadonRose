```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
B_LR_lima = pd.read_excel(os.path.join("data_procesada","B_LR_TC_G2.xlsx"))
C_LR_lima = pd.read_excel(os.path.join("data_procesada","C_LR_TC_G2.xlsx"))
D_LR_lima = pd.read_excel(os.path.join("data_procesada","D_LR_TC_G2.xlsx"))
E_LR_lima = pd.read_excel(os.path.join("data_procesada","E_LR_TC_G2.xlsx"))
```

```python
B_LR_L = B_LR_lima.copy()
C_LR_L = C_LR_lima.copy()
D_LR_L = D_LR_lima.copy()
E_LR_L = E_LR_lima.copy()
```

```python
B_LR_L = B_LR_L.iloc[1:,[0,10,11]]
B_LR_L.columns=B_LR_L.columns.str.replace(' ','_')

C_LR_L = C_LR_L.iloc[1:,[0,10,11]]
C_LR_L.columns=C_LR_L.columns.str.replace(' ','_')

D_LR_L = D_LR_L.iloc[1:,[0,10,11]]
D_LR_L.columns=D_LR_L.columns.str.replace(' ','_')

E_LR_L = E_LR_L.iloc[1:,[0,10,11]]
E_LR_L.columns=E_LR_L.columns.str.replace(' ','_')
```

```python
B_LR_L.rename(columns={'Unnamed:_0':'codigo','Valores_procesados_con_Hoja_de_calculo':'B_tr/cm2','Unnamed:_11':'B_uncert'},inplace=True)
C_LR_L.rename(columns={'Unnamed:_0':'codigo','Valores_procesados_con_Hoja_de_calculo':'C_tr/cm2','Unnamed:_11':'C_uncert'},inplace=True)
D_LR_L.rename(columns={'Unnamed:_0':'codigo','Valores_procesados_con_Hoja_de_calculo':'D_tr/cm2','Unnamed:_11':'D_uncert'},inplace=True)
E_LR_L.rename(columns={'Unnamed:_0':'codigo','Valores_procesados_con_Hoja_de_calculo':'E_tr/cm2','Unnamed:_11':'E_uncert'},inplace=True)
E_LR_L
```

```python
B_LR_L.dropna(subset=[B_LR_L.columns[0]],inplace=True)
B_LR_L = B_LR_L.reset_index(drop=True)

C_LR_L.dropna(subset=[C_LR_L.columns[0]],inplace=True)
C_LR_L = C_LR_L.reset_index(drop=True)

D_LR_L.dropna(subset=[D_LR_L.columns[0]],inplace=True)
D_LR_L = D_LR_L.reset_index(drop=True)

E_LR_L.dropna(subset=[E_LR_L.columns[0]],inplace=True)
E_LR_L = E_LR_L.reset_index(drop=True)
```

```python
def incertidumbre_media(col):
    return np.sqrt(np.mean(np.square(col)))
```

```python
B_TC2_uncertMean = TC2_L.iloc[:,2::2].apply(incertidumbre_media).values[0]
C_TC2_uncertMean = TC2_L.iloc[:,2::2].apply(incertidumbre_media).values[1]
D_TC2_uncertMean = TC2_L.iloc[:,2::2].apply(incertidumbre_media).values[2]
E_TC2_uncertMean = TC2_L.iloc[:,2::2].apply(incertidumbre_media).values[3]
```

```python
TC2_un_temp = np.array([B_TC2_uncertMean,C_TC2_uncertMean,D_TC2_uncertMean,E_TC2_uncertMean])
TC2_un_temp
```

```python
plt.bar(['9/06/2022', '23/06/2022', '7/07/2022', '21/07/2022'], TC2_temp, color='skyblue', alpha=0.7)
plt.errorbar(['9/06/2022', '23/06/2022', '7/07/2022', '21/07/2022'], TC2_temp, yerr=TC2_un_temp, fmt='o', capsize=5, label='Datos con incertidumbre',color='darkblue')
plt.ylim(bottom=0)

plt.xlabel('')
plt.ylabel('Tr/cm2_mean')
plt.title("Tr/cm2 promedio de Rn-222 de monitores TC")
plt.show()
```
